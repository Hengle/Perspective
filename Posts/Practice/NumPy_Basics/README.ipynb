# Machine Learning

## Basics

### 数据

#### Terms

- 数据集（Data Set）
- 样本（Sample）
- 特征（Feature）：通常用 $X$ 表示（通常是矩阵）
    - 第 $i$ 个样本行：$X^{(i)}$（特征向量）
    - 第 $i$ 个样本第 $j$ 个特征值：$X^{(i)}_{1}$
- 标记（Label）：通常用 $y$ 表示（向量）
    - 第 $i$ 个样本的标记：$y^{(i)}$
- 特征空间（Feature Space）



- 分类任务本质就是在特征空间切分
- 在高维空间同理
- 特征可以很抽象
    - 图像，每一个像素点都是特征点
    - MNIST：28 * 28 的图像即有 28 * 28 = 784 个特征
    - 如果是彩色图像特征更多

#### 基本任务

- 监督学习
    - 给机器的训练数据拥有「标记」或者「答案」
    - 分类
        - 二分类
        - 多分类
        - 多标签分类
    - 回归（结果是一个连续数字的值，而非一个类别）
    - 算法：k 近邻，线性回归，多项式回归，逻辑回归，SVM，决策树，随机森林 
- 非监督学习
    - 给机器的训练数据没有任何「标记」或者「答案」
    - 对没有「标记」的数据进行分类 - 聚类分析
    - 意义
        - 对数据进行降维处理
            - 特征提取：信用卡的信用评级和人的胖瘦无关？
            - 特征压缩：PCA
            - 意义：方便可视化
        - 异常检测
- 半监督学习
    - 一部分数据有「标记」或者「答案」，另一部分数据没有
    - 更常见于：各种原因产生的标记缺失
    - 通常都先使用无监督学习手段对数据做处理，之后使用监督学习手段做模型的训练和预测
- 增强学习
    - 根据周围环境的情况，采取行动，根据采取行动的结果，学习行动方式
    - AlphaGo


- 在线学习 Online Learning
    - 及时反映新的环境变化
    - 需要加强对数据进行监控（异常检测）
- 批量学习 Batch Learning
    - 又称离线学习 Offline Learning
    - 优点：简单
    - 如何适应环境变化？
        - 定时重新批量学习

- 参数学习（Parametric Learning）
    - 一旦学到了参数，就不再需要原有的数据集
- 非参数学习（Nonparametric Learning）
    - 不对模型进行过多假设
    - 非参数不等于没参数
